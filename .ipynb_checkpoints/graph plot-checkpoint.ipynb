```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.load('data.npy', allow_pickle=True)
step = np.linspace(1, 200, 200)
param = data[0]
loss = data[1]
```

```python
df = pd.DataFrame({'step': step, 'loss': loss})

df['std'] = df['loss'].rolling(50).std()
df['mean'] = df['loss'].rolling(50).mean()

df['minbox'] = df['mean'] - df['std']/5
df['maxbox'] = df['mean'] + df['std']/5

bx = df.plot(x='step', y = 'mean', figsize=(9, 6), label='# of molecules = 5 ')
plt.fill_between(df['step'], df['minbox'], df['maxbox'], color='blue', alpha = 0.2)
#plt.show()

plt.xlabel('num of iteration', labelpad=10)
plt.ylabel('Loss Function', labelpad=10)

plt.savefig('good.png',dpi=300)
```
